Creating the .xyn file is using the .pli to obs command from interactor. But those points then need to be converted to the boundary points for the regional model grid. So we do that here.  **Don't forget to delete the -999 point in resulting .xyn file from interactor** 

```python
j=0.005
file='Canada_model_tests_interactor/fes10km_5e-2_obs.xyn'
# with open(file,'r') as f:
#     lines=f.readlines()

#getting vectors of coordinates.
data = np.loadtxt(file, skiprows=0,usecols=(0,1),dtype='float')
data1=np.where(data==-47.104375,(-47.104375-j),data)
data1=np.where(data==50.385625,(50.385625+j),data1)
data1=np.where(data==-59.99500 ,(-59.99500 -j),data1)
data1=np.where(data==83.208125,(83.208125-j),data1)
data1=np.where(data==-157.505000,(-157.505000+j),data1)
obsname=np.loadtxt(file,usecols=(2),dtype='str')
```

```python
vec=np.hstack((data1,obsname.reshape(len(obsname),1)))
print(vec.shape)
print(vec)
```

```python
np.savetxt('Canada_model_tests_interactor/model10km_5e-2_obs.xyn',vec,fmt='%s',delimiter="\t")
```
