```python
import numpy as np
import os
```

```python
file='Canada_model/only_waterboundar.pol'
with open(file,'r') as f:
    lines=f.readlines()

print(lines)
```

```python
#getting vectors of coordinates.
data = np.loadtxt(file, max_rows=33, skiprows=2, dtype='float')
data1=np.where(data==-47.109375,(-47.109375+0.06),data)
data1=np.where(data==50.390625,(50.390625-0.06),data1)
# print((data1))
# print(data)
```

```python
data2 = np.loadtxt(file, skiprows=37, dtype='float')
data22=np.where(data2==-60.00000,(-60.00000+0.06),data2)
data22=np.where(data2==83.203125,(83.203125+0.06),data22)
data22=np.where(data2==-157.5,(-157.5-0.06),data22)
print(data22)
```

```python
for line in lines:
    if len(line)==13:
        print(line)
```

```python
newfile='Canada_model/boundary.pol'
nfileobj=open(newfile,'w')
i=0
for line in lines:
    if line[:] =='L1\n':
        nfileobj.write('right_boundary\n')
        j=0
        for line in lines[i+1:len(data1[:,0])+2]:
            if len(line)==13:
                nfileobj.write(line)
            else:
                nfileobj.write('     '+str(data1[j,0])+'      '+str(data1[j,1])+'\n')
                j+=1
    elif line[:]=='L2\n':
        nfileobj.write('left_boundary\n') 
        j=0
        for line in lines[i+1:]:
            if len(line)==13:
                nfileobj.write(line)
            else:
                nfileobj.write('     '+str(data22[j,0])+'      '+str(data22[j,1])+'\n')
                j+=1

    i+=1
    
nfileobj.close()
```
